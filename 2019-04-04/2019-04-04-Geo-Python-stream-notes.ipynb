## Plotting random points with random colors (Problem 1 of Ex. 7)

```python
# Load the modules we need
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
# Make some random points
x = np.random.rand(1000)
y = np.random.rand(1000)
colors = np.random.rand(1000)
```

```python
# Make our colorful scatter plot
plt.scatter(x=x, y=y, c=colors, cmap='jet')
```

`plt.plot()` can be used as an alternative to `plt.scatter()`, but you can't change the marker colors indivudually.

```python
plt.plot(x, y, color='#FF6666', marker='D', linestyle=':', markersize=2)
```
